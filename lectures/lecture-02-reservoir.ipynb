# Math 527 lecture 2: general solution versus solution of initial value problem reservoir problem

In lecture we showed that water drains out a cylindrical reservoir according to the differential equation

\begin{align*}
\frac{dy}{dt} = a y
\end{align*}

where $y(t)$ is the height of the water above the spigot. We used separation of variables to find the solution of the differential equation.

\begin{align*}
y(t) = c e^{at}
\end{align*}

The constant $a$ is determined by properties of the reservoir and the fluid, specifically $a  = - \rho g r^4/(8 \mu \ell R^2)$ where from $\rho$ is the fluid density $\rho$, $\mu$ is the fluid dynamic viscosity, $R$ is the radius of the reservoir $R$, $r$ is the radius $r$ and $\ell$ the length of the spigot, and $g$ is the acceleration of gravity. Mechanical engineers, please double-check my derivation of $a$! I've assumed the flow rate is determined by the pressure drop from laminar flow through the cylindrical spigot.

For an reservoir with radius $R = 10m$, a spigot of length $\ell = 1m$ and radius $r = 0.02m$, filled with water of density $\rho = 1000 kg/m^3$ and kinematic viscosity $\mu = 10^{-3} Ns/m^2$, and $g = 9.8 m/2^2$, we get $a \approx -0.002$. 

In this case the ODE is 
\begin{align*}
\frac{dy}{dt} = -0.002 y
\end{align*}

and the general solution is 

\begin{align*}
y(t) = c e^{-0.002 t}
\end{align*}
