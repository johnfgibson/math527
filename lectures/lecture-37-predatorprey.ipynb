```julia
using PyPlot
using ODE

 = 1;  = 2;  = 1;  = 1;
xmax = 2
ymax = 1
xgrid = linspace(0,xmax,20)
ygrid = linspace(0,ymax,20)
dxdt = [*x - *x*y for y in ygrid, x in xgrid]
dydt = [*x*y - *y for y in ygrid, x in xgrid]
quiver(xgrid, ygrid, dxdt, dydt, width=0.002)

plot([0], [0], "bo")
plot([/], [/], "bo")
xlabel("x (rabbits)")
ylabel("y (wolves)")
t = linspace(0,8,20)
plot(0*t, ymax*exp.(-*t), "b-")
plot(0.02*exp.(*t), 0*t, "r-")
xlim(-0.1, xmax)
ylim(-0.1, ymax)

function f(t,x) 
    [*x[1] - *x[1]*x[2]; *x[1]*x[2] - *x[2]]
end

t = 0:0.02:16.0
x = [1.0; 0.55]
t,x = ode45(f, x, t)
x = hcat(x...)'    

plot(x[:,1], x[:,2], "k-")

x = [0.5; 0.5]
t,x = ode45(f, x, t)
x = hcat(x...)'    

plot(x[:,1], x[:,2], "k-")
```

```julia
plot(t, x[:,1], "b", label="x, rabbits")
plot(t, x[:,2], "r", label="y, wolves")
legend()
```
