```julia
using PyPlot
using ODE

α = 1; β = 2; γ = 1; δ = 1;
xmax = 2
ymax = 1
xgrid = linspace(0,xmax,20)
ygrid = linspace(0,ymax,20)
dxdt = [α*x - β*x*y for y in ygrid, x in xgrid]
dydt = [δ*x*y - γ*y for y in ygrid, x in xgrid]
quiver(xgrid, ygrid, dxdt, dydt, width=0.002)

plot([0], [0], "bo")
plot([γ/δ], [α/β], "bo")
xlabel("x (rabbits)")
ylabel("y (wolves)")
t = linspace(0,8,20)
plot(0*t, ymax*exp.(-γ*t), "b-")
plot(0.02*exp.(α*t), 0*t, "r-")
xlim(-0.1, xmax)
ylim(-0.1, ymax)

function f(t,x) 
    [α*x[1] - β*x[1]*x[2]; δ*x[1]*x[2] - γ*x[2]]
end

t = 0:0.02:16.0
x₀ = [1.0; 0.55]
t,x = ode45(f, x₀, t)
x = hcat(x...)'    

plot(x[:,1], x[:,2], "k-")

x₀ = [0.5; 0.5]
t,x = ode45(f, x₀, t)
x = hcat(x...)'    

plot(x[:,1], x[:,2], "k-")
```

```julia
plot(t, x[:,1], "b", label="x, rabbits")
plot(t, x[:,2], "r", label="y, wolves")
legend()
```
