```julia
using PyPlot
t = linspace(0, 10, 200)
k = sqrt(16)
U(x::AbstractFloat) = ifelse(x < 0, zero(x), ifelse(x > 0, one(x), oftype(x,0.5)))

y0 = -1/3*(1-exp.(-t).*(cos.(k*t) + 1/k * sin.(k*t)))

y1 = 1/3*((exp.(-t).*(cos.(k*t) + 1/k * sin.(k*t)) - 1)  - U.(t-2).*(exp.(-(t-2)).*(cos.(k*(t-2)) + 1/k * sin.(k*(t-2))) - 1))
```

```julia
plot(t,y0,"b--")
plot(t,y1,"b-")
xlabel("t")
ylabel("y(t)")
```
